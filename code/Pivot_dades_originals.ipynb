```python
import pandas as pd
import numpy as np
pd.options.display.max_rows = 1000
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

df = pd.read_excel('pax_all_agreements_data.xlsx')

df.head()
```

```python
df['Con_normalized'] = [cell.replace('(', '').replace(')', '').strip() for cell in df['Con']]
df['Con_normalized'].value_counts()
```

```python
# Aqui afegim el nou ...
_
df['_groups'] = 0 
df['_gender']  = 0 
df['_state_definition']  = 0 
df['_governance']  = 0 
df['_powersharing'] = 0
df['_human_rights']  = 0
df['_justice']  = 0 
df['_socio_economic']  = 0 
df['_land_property']  = 0
df['_security']  = 0 
df['_transitional_justice']  = 0
df['_implementation']  = 0

columnes = {
	"_groups" : ['GCh','GChRhet','GChAntid','GChSubs','GChOth','GDis','GDisRhet','GDisAntid','GDisSubs',
               'GDisOth','GAge','GAgeRhet','GAgeAntid','GAgeSubs','GAgeOth','GMig','GMigRhet','GMigAntid',
               'GMigSubs','GMigOth','GRa','GRaRhet','GRaAntid','GRaSubs','GRaOth','GRe','GReRhet','GReAntid',
               'GReSubs','GReOth','GInd','GIndRhet','GIndAntid','GIndSubs','GIndOth','GOth','GOthRhet','GOthAntid',
               'GOthSubs','GOthOth','GRef','GRefRhet','GRefSubs','GRefOth','GSoc','GSocRhet','GSocAntid','GSocSubs',
               'GSocOth'],
    "_gender" : ['GeWom','GeMe','GeMeNu','GeMeOth','GeLgbti','GeLgbtiPos','GeLgbtiNeg','GeFa'],
	"_state_definition" : ['StDef','StGen','StCon','StSd','StRef','StSym','StInd','StUni','StBor','StXbor'],
	"_governance" :  ['Pol','PolGen','PolNewInd','PolNewTemp','ConRen','Cons','Ele','ElecComm','PolPar',
                    'PolParTrans','PolParOth','Civso','Tral','Pubad'],
    "_powersharing" : ['Polps','PpsSt','PpsSub','PpsEx','PpsOro','PpsOthPr','PpsVet','PpsAut','PpsInt','PpsOth','Terps',
                     'TpsSub','TpsLoc','TpsAut','TpsOth','Eps','EpsRes','EpsFis','EpsOth','Mps','MpsMe','MpsJt','MpsPro','MpsOth'],
    "_human_rights" : ['HrGen','EqGen','HrDem','Prot','ProtCiv','ProtGrp','ProtLgl','ProtOth','HrFra','HrfSp','HrfBor','HrfTinc','HrfOth','HrCp',
                'CprLife','CprTort','CprEq','CprSlav','CprLib','CprDet','CprFmov','CprFspe','CprFass','CprTria','CprPriv','CprVote',
                'CprReli','CprOth','HrSec','SerProp','SerWork','SerHeal','SerEdu','SerStdl','SerShel','SerSs','SerCult','SerOth','HrNi',
                'HrNiMe','HrNiNe','HrNiOth','HrIi','HrIiMon','HrIiBod','HrIiOth','HrMob','HrDet','Med','MedGov','MedSubs',
                'MedLog','MedOth','HrCit','CitGen','CitRights','CitDef','CitOth'],
    "_justice" :  ['JusCr','JusCrSp','JusCrSys','JusCrPow','JusEm','JusJu','JusPri','JusTra'],
    "_socio_economic" : ['Dev','DevSoc','DevHum','DevInfra','NEC','NatRes','IntFu','Bus','Tax',
                       'TaxPo','TaxRef','TaxOth','Ban','CenBan','BanPers','BanInt','BanXb'],
    "_land_property": ['LaRef','LaRefMan','LaRefRet','LaRefOth','LaNom','LaCH','LaCHTa','LaCHIt','LaCHPro','LaCHOth',
                      'LaEn','Wat'],
    "_security" : ['SsrGua','Ce','CeProv','CeGen','SsrPol','SsrArm','SsrDdr','DdrDemil','DdrProg','SsrInt',
                 'SsrPsf','SsrFf','Cor','SsrCrOcr','SsrDrugs','Terr'],
    "_transitional_justice": ['TjGen','TjAm','TjAmPro','TjSan','TjPower','TjAmBan','TjCou','TjJaNc','TjJaIc','TjMech',
                        'TjPrire','TjVet','TjVic','TjMis','TjRep','TjRSym','TjRMa','TjNR'],
    "_implementation" :  ['ImUN','ImOth','ImRef','ImPK','ImE','ImSrc']
}


def calcularColumna(nom_columna_calculada):
    a = []
    for index, row in df.iterrows():         
        columnes_df = columnes.get(nom_columna_calculada)
        valor_nova_columna = 0
        for columna in columnes_df:
            if (row[columna]!=0):
                valor_nova_columna = 1
                break;
        a.append(valor_nova_columna) 
    print("nova_columna: {}, items: {}".format(nom_columna_calculada, len(a)))    
    return a


def calcularColumnaPromig(nom_columna_calculada):
    a = []
    for index, row in df.iterrows():         
        columnes_df = columnes.get(nom_columna_calculada)
        valor_nova_columna = 0
        for columna in columnes_df:
            if (row[columna]!=0):
                valor_nova_columna += 1
        a.append(valor_nova_columna/len(columnes_df)) 
    print("nova_columna: {}, items: {}".format(nom_columna_calculada, len(a)))    
    return a

for nova_columna, columnes_existents in columnes.items():
    valors_nova_columna = calcularColumna(nova_columna)
    df[nova_columna] = np.array(valors_nova_columna)   


df_conflict_types = pd.DataFrame()
cols = list(columnes.keys())
cols.insert(0, "AgtId")

df_conflict_types = pd.DataFrame(columns=cols)
df_conflict_types["AgtId"] = df["AgtId"]

for c in columnes.keys():
    df_conflict_types[c] = df[c]


df_conflict_types_melted = pd.melt(df_conflict_types, id_vars=["AgtId"], value_vars=list(columnes.keys()))
df_conflict_types_melted = df_conflict_types_melted[~(df_conflict_types_melted == 0).any(axis=1)]
df_conflict_types_melted = df_conflict_types_melted.drop(['value'], axis = 1)

df_conflict_types_melted= df_conflict_types_melted.sort_values(by=['AgtId'])
df_conflict_types_melted.to_excel("pivot_groups_per_tractat.xlsx", index=False)
```

```python
unique_con = sorted(df['Con_normalized'].unique())
unique_con_slash = [token for token in unique_con if "/" in token]


land_list = set()

for token in unique_con_slash:
    token_splited = token.split('/')
    for word in token_splited:
        land_list.add(word)


cols = list(land_list)
cols.insert(0, "AgtId")
cols.insert(1, "Con_normalized")

df_2 = pd.DataFrame(columns=cols)
df_2["AgtId"] = df["AgtId"]
df_2["Con_normalized"] = df["Con_normalized"]
```

```python
for index, row in df_2.iterrows():
    token_splited = row["Con_normalized"].split('/')
    for word in token_splited:
        df_2.loc[index, word] = word
   
```

```python
columns_2 =  [col for col in df_2.columns if col not in ('AgtId','Con_normalized')]
```

```python
df_melted = pd.melt(df_2, id_vars=["AgtId", "Con_normalized"], value_vars=columns_2)
print(df_melted)
```

```python
df_melted_nan = df_melted[df_melted.value.notnull()]
df_melted_nan = df_melted_nan.drop(['value'], axis = 1)
df_melted_nan = df_melted_nan.sort_values(by=['AgtId'])
```

```python
df_country_code = pd.read_excel('UNSD — Methodology.xlsx')

col_list = ['Country or Area', 'ISO-alpha3 Code']

df_country_code = df_country_code[col_list]


df_country_code.head()

df_merge_col = pd.merge(df_melted_nan, df_country_code, left_on='variable', right_on='Country or Area', how='outer')
df_merge_col = df_merge_col.sort_values(by=['AgtId'])
df_merge_col.to_excel("pivot_paisos_per_tractat.xlsx", index=False)
```
